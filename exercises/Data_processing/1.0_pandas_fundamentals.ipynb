 - d) Filter out the five smallest cities. (*)

 - e) Use the DataFrame to calculate Sweden's population in 2019 and 2020. (*)

```python
population_2019 = df["Folkmängd 2019"].sum()
population_2020 = df["Folkmängd 2020"].sum()


print(f"The popultion of sweden in 2019 was {population_2019}.")
print(f"The popultion of sweden in 2020 was {population_2020}. ")
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns



fig, axes = plt.subplots(1,2, dpi = 120 , figsize = (16,4))


dfs = [ Top_city_population , smalles_city]

x_column = ["Folkmängd 2020" , "Folkmängd 2020" ]
titles = [ "Top_city_population" , "smalles_city"]

for ax , data, title , x_columns in zip(axes , dfs, titles, x_column):
    sns.barplot(data , x=x_column, y= "Population", ax = ax)
    ax.set(titles = titles)
    ax.set_xticklabels(ax.get_xticklabels(), rotation = 90)


plt.show()
```
