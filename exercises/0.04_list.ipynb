### 0. Dice rolls (*)
- Simulate 10 dice rolls and append the rolls to a list or use list comprehension.

  - a)   sort the list in ascending order (*)


 

```python
import random

roll_dics = [random.randint(1,6) for i in range(10)]

roll_dics_accending = sorted(roll_dics )

print(roll_dics)
print(roll_dics_accending)
```

- b)   sort the list in descending order (*)

```python
import random

roll_dics = [random.randint(1,6) for i in range(10)]

roll_dics_descending = sorted(roll_dics , reverse=True)

print(roll_dics)
print(roll_dics_descending)
```

```python
import random

roll_dics = [random.randint(1,6) for i in range(10)]

min_num = min(roll_dics)
max_num = max(roll_dics)

print(roll_dics)
print(f"Min number {min_num}")
print(f"max number {max_num}")
```

2. Squares (*)
  - a)   Use list comprehension to create a list of squares from -10 to 10 (*)

  - b)   Plot this list using matplotlib. (*)

```python
sqr= [x**2 for x in range (-10,11)]
print(sqr)
```

```python
import matplotlib.pyplot as plt

sqr= [x**2 for x in range (-10,11)]

plt.plot(range(-10,11),sqr , label = "square of X")

plt.xlabel("x")
plt.ylabel("y")
plt.title("square of X")

plt.legend()
plt.show()
```
