# B) Ta reda på antalet som fått F i matematik totalt, pojkar respektive flickor.

```python
df_skola = pd.read_excel("../Uppgift_0/riket2023_åk9_np.xlsx", sheet_name= "Matematik" , skiprows=8)

column_rename = {df_skola.columns[2]: "Totalt (A-F)" , df_skola.columns[3]: "Flickor (A-F)" , df_skola.columns[4]: "Pojkar (A-F)" , df_skola.columns[5]: "Totalt (A-E)" , df_skola.columns[6]: "Flickor (A-E)" , df_skola.columns[7]: "Pojkar (A-E)", df_skola.columns[8]: "Totalt (Poäng)" , df_skola.columns[9]: "Flickor (Poäng)" , df_skola.columns[10]: "Pojkar (Poäng)"}

df_skola = df_skola.rename(columns=column_rename)

df_skola_mate = df_skola

df_skola_mate
```

```python
droping_column = [8 , 9,10]
df_fått_F = df_skola_mate.drop(df_skola_mate.columns[droping_column], axis=1)
df_fått_F
```

```python

df_fått_F = pd.DataFrame(df_fått_F)
df_fått_F["Totalt(F)"] = pd.Series(dtype=float)
df_fått_F["Flickor(F)"]= pd.Series(dtype=float)
df_fått_F["Pojkar(F)"] = pd.Series(dtype=float)
df_fått_F
```

```python
df_fått_F['Totalt (A-E)'] = df_fått_F['Totalt (A-E)'].replace({'~': ''}, regex=True)
df_fått_F['Flickor (A-E)'] = df_fått_F['Flickor (A-E)'].replace({'~': ''}, regex=True)
df_fått_F['Pojkar (A-E)'] = df_fått_F['Pojkar (A-E)'].replace({'~': ''}, regex=True)
df_fått_F
```

```python

# Assuming df_fått_F is your DataFrame
df_fått_F['Totalt (A-F)'] = pd.to_numeric(df_fått_F['Totalt (A-F)'])
df_fått_F['Totalt (A-E)'] = pd.to_numeric(df_fått_F['Totalt (A-E)']) / 100 * 100

df_fått_F['Flickor (A-F)'] = pd.to_numeric(df_fått_F['Flickor (A-F)'])
df_fått_F['Flickor (A-E)'] = pd.to_numeric(df_fått_F['Flickor (A-E)']) / 100 * 100

df_fått_F['Pojkar (A-F)'] = pd.to_numeric(df_fått_F['Pojkar (A-F)'])
df_fått_F['Pojkar (A-E)'] = pd.to_numeric(df_fått_F['Pojkar (A-E)']) / 100 * 100



# Print the DataFrame
df_fått_F
```

```python

df_fått_F['Totalt(F)'] = df_fått_F['Totalt (A-F)'] * (100 - df_fått_F['Totalt (A-E)']) /100
df_fått_F['Flickor(F)'] = df_fått_F['Flickor (A-F)'] * (100 - df_fått_F['Flickor (A-E)']) / 100
df_fått_F['Pojkar(F)'] = df_fått_F['Pojkar (A-F)'] * (100 - df_fått_F['Pojkar (A-E)']) / 100

df_fått_F = df_fått_F
df_fått_F
```

# D) Rita fyra olika stapeldiagram i subplots på totala poängen för de olika ämnena för de olika huvudmännen.

```python

# Extracting the first values for each subject
values_samtliga = pd.DataFrame({
    'Subject': ['Engelska', 'SSAS', 'Mate', 'Svenska'],
    'Samtliga': [df_skola_engelska.loc[df_skola_engelska.index[0], 'Totalt (Poäng)'],
                df_skola_SSAS.loc[df_skola_SSAS.index[0], 'Totalt (Poäng)'],
                df_skola_mate.loc[df_skola_mate.index[0], 'Totalt (Poäng)'],
                df_skola_svenska.loc[df_skola_svenska.index[0], 'Totalt (Poäng)']]
})


values_kommunal = pd.DataFrame({ 
    'Subject': ['Engelska', 'SSAS', 'Mate', 'Svenska'],
    'Kommunal': [
    df_skola_engelska.loc[df_skola_engelska.index[1], 'Totalt (Poäng)'],
    df_skola_SSAS.loc[df_skola_SSAS.index[1], 'Totalt (Poäng)'],
    df_skola_mate.loc[df_skola_mate.index[1], 'Totalt (Poäng)'],
    df_skola_svenska.loc[df_skola_svenska.index[1], 'Totalt (Poäng)']]

})


values_enskild = pd.DataFrame({
    'Subject': ['Engelska', 'SSAS', 'Mate', 'Svenska'],
    'Enskild' : [df_skola_engelska.loc[df_skola_engelska.index[2], 'Totalt (Poäng)'],
                df_skola_SSAS.loc[df_skola_SSAS.index[2], 'Totalt (Poäng)'],
                df_skola_mate.loc[df_skola_mate.index[2], 'Totalt (Poäng)'],
                df_skola_svenska.loc[df_skola_svenska.index[2], 'Totalt (Poäng)']]

})

values_skolverket = pd.DataFrame({
    'Subject': ['Engelska', 'SSAS', 'Mate', 'Svenska'],
    'Skolverket' : [df_skola_engelska.loc[df_skola_engelska.index[3], 'Totalt (Poäng)'],
                np.nan, #df_skola_SSAS.loc[df_skola_SSAS.index[3], 'Totalt (Poäng)'],
                df_skola_mate.loc[df_skola_mate.index[3], 'Totalt (Poäng)'],
                df_skola_svenska.loc[df_skola_svenska.index[3], 'Totalt (Poäng)']]

})
merged_values = pd.concat([values_samtliga.set_index('Subject'), 
                        values_kommunal.set_index('Subject'), 
                        values_enskild.set_index('Subject'), 
                        values_skolverket.set_index('Subject')], axis=1)

print(merged_values)
```

```python
import matplotlib.pyplot as plt

# Assuming merged_values is your merged DataFrame

# Create subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

# Plot 'Samtliga' in the first subplot
merged_values['Samtliga'].plot(kind='bar', ax=axes[0, 0], title='Samtliga')

# Plot 'Kommunal' in the second subplot
merged_values['Kommunal'].plot(kind='bar', ax=axes[0, 1], title='Kommunal')

# Plot 'Enskild' in the third subplot
merged_values['Enskild'].plot(kind='bar', ax=axes[1, 0], title='Enskild')

# Plot 'Skolverket' in the fourth subplot
merged_values['Skolverket'].plot(kind='bar', ax=axes[1, 1], title='Skolverket')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
