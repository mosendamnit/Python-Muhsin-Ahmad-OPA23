# B) Ta reda på antalet som fått F i matematik totalt, pojkar respektive flickor.

```python
df_skola = pd.read_excel("../Uppgift_0/riket2023_åk9_np.xlsx", sheet_name= "Matematik" , skiprows=8)

column_rename = {df_skola.columns[2]: "Totalt (A-F)" , df_skola.columns[3]: "Flickor (A-F)" , df_skola.columns[4]: "Pojkar (A-F)" , df_skola.columns[5]: "Totalt (A-E)" , df_skola.columns[6]: "Flickor (A-E)" , df_skola.columns[7]: "Pojkar (A-E)", df_skola.columns[8]: "Totalt (Poäng)" , df_skola.columns[9]: "Flickor (Poäng)" , df_skola.columns[10]: "Pojkar (Poäng)"}

df_skola = df_skola.rename(columns=column_rename)

df_skola_mate = df_skola

df_skola_mate
```

```python
droping_column = [8 , 9,10]
df_fått_F = df_skola_mate.drop(df_skola_mate.columns[droping_column], axis=1)
df_fått_F
```

```python
import numpy as np

df_fått_F = pd.DataFrame(df_fått_F)
df_fått_F["Totalt(F)"] = pd.Series(dtype=float)
df_fått_F["Flickor(F)"]= pd.Series(dtype=float)
df_fått_F["Pojkar(F)"] = pd.Series(dtype=float)
df_fått_F
```

```python
df_fått_F['Totalt (A-E)'] = df_fått_F['Totalt (A-E)'].replace({'~': ''}, regex=True)
df_fått_F['Flickor (A-E)'] = df_fått_F['Flickor (A-E)'].replace({'~': ''}, regex=True)
df_fått_F['Pojkar (A-E)'] = df_fått_F['Pojkar (A-E)'].replace({'~': ''}, regex=True)
df_fått_F
```

```python
import pandas as pd

# Assuming df_fått_F is your DataFrame
df_fått_F['Totalt (A-F)'] = pd.to_numeric(df_fått_F['Totalt (A-F)'])
df_fått_F['Totalt (A-E)'] = pd.to_numeric(df_fått_F['Totalt (A-E)']) / 100 * 100

df_fått_F['Flickor (A-F)'] = pd.to_numeric(df_fått_F['Flickor (A-F)'])
df_fått_F['Flickor (A-E)'] = pd.to_numeric(df_fått_F['Flickor (A-E)']) / 100 * 100

df_fått_F['Pojkar (A-F)'] = pd.to_numeric(df_fått_F['Pojkar (A-F)'])
df_fått_F['Pojkar (A-E)'] = pd.to_numeric(df_fått_F['Pojkar (A-E)']) / 100 * 100



# Print the DataFrame
df_fått_F
```

```python
import pandas as pd

df_fått_F['Totalt(F)'] = df_fått_F['Totalt (A-F)'] * (100 - df_fått_F['Totalt (A-E)']) /100
df_fått_F['Flickor(F)'] = df_fått_F['Flickor (A-F)'] * (100 - df_fått_F['Flickor (A-E)']) / 100
df_fått_F['Pojkar(F)'] = df_fått_F['Pojkar (A-F)'] * (100 - df_fått_F['Pojkar (A-E)']) / 100

df_fått_F = df_fått_F
df_fått_F
```
